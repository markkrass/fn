## Step 1: Import data from previous notebook

```python
data = pd.read_csv("sampledat.csv", na_values="NA", parse_dates=True)
del data['X']
data.replace(to_replace="Robinson Huron 1850", value="Robinson-Huron Treaty", inplace=True)

print data.columns
data.head(2)
```

## Step 2: Group by Treaty and by Primary Tribal Organization [PTO]

```python
#by treaty
dftre = data.groupby('treaty')
# by pto
dfpto = data.groupby('tribal_org')

# Get counts for each
counttre = dftre.agg('count')['name']
countpto = dfpto.agg('count')['name']

# Get proportion that fail to report URLs

NAurl_tre = (counttre - dftre.agg('count')['url'])/counttre
NAurl_pto = (countpto - dfpto.agg('count')['url'])/countpto
```

```python
# Get total population
SigPoptre = dftre.agg('sum')['on_reserve_pop']
SigPoppto = dfpto.agg('sum')['on_reserve_pop']

# Proportion of FN on-reserve population by cut
PropPoptre = SigPoptre / np.sum(data['on_reserve_pop'])
PropPoppto = SigPoppto / np.sum(data['on_reserve_pop'])

# Median on-reserve population

med_pop_tre = dftre.agg('median')['on_reserve_pop']
med_pop_pto = dfpto.agg('median')['on_reserve_pop']
```

```python
averages_tre = dftre.agg('mean')[['median_hh_income','lfpartic', 'hsdip_15plus', 'unemployment','pct_young','FNlang_know']]
averages_pto = dfpto.agg('mean')[['median_hh_income','lfpartic', 'hsdip_15plus', 'unemployment','pct_young','FNlang_know']]
```

## Step 3: Plot in MatPlotLib

```python
def plot_by(by_frame, title, variable=False, colors=more_colors, baseline=False):
    """
    A little utility for helping to plot all of these things....
    Note that if by_frame is an np.array, a variable is not necessary
    but if by_frame is a dataframe callable dict-style (x['y']), then
    a str containing the name of the desired column is needed in variable. 
    
    Baseline should be a dict.
    """
    if variable:
        data_col = np.array(by_frame[variable])
    if not variable:
        data_col = np.array(by_frame)
    if baseline:
        labs = np.array(by_frame.index)
        labs = np.append(labs, 'Ontario')
        plt.figure()
        plt.bar(left=range(len(by_frame.index)), height=data_col, 
                color=colors, hold=True)
        plt.bar(left=8, height=np.array(baseline[variable]), color="black", alpha=0.2)
        plt.title(title)
        xticks_locs, xticks_labels = plt.xticks([a+ 0.5 for a in range(len(counttre)+1)], labs)
        plt.setp(xticks_labels, rotation=90, fontsize=12)
        x1,x2,y1,y2 = plt.axis()
        plt.axis((x1,9,y1,(np.max(np.append(data_col, baseline[variable]))*1.1)))
        plt.show()
    if not baseline:
        plt.figure()
        plt.bar(left=range(len(by_frame.index)), height=np.array(data_col), color=more_colors)
        plt.title(title)
        xticks_locs, xticks_labels = plt.xticks([a+ 0.5 for a in range(len(by_frame))], np.array(by_frame.index))
        plt.setp(xticks_labels, rotation=90, fontsize=12)
        x1,x2,y1,y2 = plt.axis()
        plt.axis((x1,x2,0,np.max(data_col)*1.1))
        plt.show()
    return
                        
```
