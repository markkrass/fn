```python
%matplotlib inline

import matplotlib.pyplot as plt

pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 30)

# set some nicer defaults for matplotlib
from matplotlib import rcParams

#these colors come from colorbrewer2.org. Each is an RGB triplet
dark2_colors = [(0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843),
                (0.4, 0.4, 0.4)]

more_colors = ['#b2182b', '#d6604d', '#f4a582', '#fddbc7', '#d1e5f0', '#92c5de', '#4393c3', '#2166ac']

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.grid'] = False
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'none'



def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecessary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
        
def plot_by(by_frame, title, variable=False, colors=more_colors, baseline=False, sidelegend=False):
    """
    A little utility for helping to plot all of these things....
    Note that if by_frame is an np.array, a variable is not necessary
    but if by_frame is a dataframe callable dict-style (x['y']), then
    a str containing the name of the desired column is needed in variable. 
    
    Baseline should be a dict.
    """
    if variable:
        data_col = np.array(by_frame[variable])
    if not variable:
        data_col = np.array(by_frame)
    orignum = len(data_col)
    if baseline:
        if not sidelegend:
            labs = np.array(by_frame.index)
            labs = np.append(labs, 'Ontario')
            plt.figure()
            plt.bar(left=range(orignum), height=data_col, 
                    color=colors, hold=True)
            plt.bar(left=orignum, height=np.array(baseline[variable]), color="black", alpha=0.2)
            plt.title(title)
            xticks_locs, xticks_labels = plt.xticks([a+ 0.5 for a in range(orignum+1)], labs)
            plt.setp(xticks_labels, rotation=90, fontsize=12)
            x1,x2,y1,y2 = plt.axis()
            plt.axis((x1,orignum+1,y1,(np.max(np.append(data_col, baseline[variable]))*1.1)))
            plt.show()
        if sidelegend:
            plt.figure()
            for i in range(len(by_frame.index)):
                plt.bar(left=range(orignum)[i], height=data_col[i], 
                    color=colors[i], label=np.array(by_frame.index)[i])
            plt.bar(left=orignum, height=np.array(baseline[variable]), color="black", alpha=0.2, label="Ontario")
            plt.title(title)
            plt.xticks([])
            x1,x2,y1,y2 = plt.axis()
            plt.axis((x1,orignum+1,y1,(np.max(np.append(data_col, baseline[variable]))*1.1)))
            plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., prop={'size':15})
            remove_border()
            plt.show()
    if not baseline:
        if not sidelegend:
            plt.figure()
            plt.bar(left=range(len(by_frame.index)), height=np.array(data_col), color=more_colors)
            plt.title(title)
            xticks_locs, xticks_labels = plt.xticks([a+ 0.5 for a in range(len(by_frame))], np.array(by_frame.index))
            plt.setp(xticks_labels, rotation=90, fontsize=12)
            x1,x2,y1,y2 = plt.axis()
            plt.axis((x1,x2,0,np.max(data_col)*1.1))
            plt.show()
        if sidelegend:
            plt.figure()
            for i in range(len(by_frame.index)):
                plt.bar(left=range(len(by_frame.index))[i], height=data_col[i], 
                    color=colors[i], label=np.array(by_frame.index)[i])
            plt.title(title)
            plt.xticks([])
            x1,x2,y1,y2 = plt.axis()
            plt.axis((x1,orignum,y1,(np.max(data_col)*1.1)))
            plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., prop={'size':15})
            remove_border()
            plt.show()
            
    return
                        
```

## (1) Histograms

```python
def make_hist(df, var, title, col, bins=10, alpha=0.7, density=True, baseline=False, baseline_col=False, legend_loc="upper left"):
    hist, edges = np.histogram(df[var].dropna(), density=density, bins=bins)
    plt.figure()
    if not baseline:
        plt.bar(left=edges[:-1], height=hist, width = (edges[2]-edges[1]), color=col, alpha=1)
    if baseline:
        plt.bar(left=edges[:-1], height=hist, width = (edges[2]-edges[1]), color=col, alpha=alpha, hold=True, label='FN Communities')
        plt.bar(left=baseline[var], height=np.max(hist), width=(0.5)*(edges[2]-edges[1]), color=baseline_col, alpha=alpha, label='Ontario baseline')
        x1,x2,y1,y2 = plt.axis()
        plt.axis((x1,(np.max(np.append(df[var].dropna(), baseline[var]))*1.1),y1,y2))
        plt.legend(loc=legend_loc, prop={'size':13})
    plt.title(title)
    if density:
        plt.ylabel('Density')
        frame1 = plt.gca()
        frame1.axes.get_yaxis().set_ticks([])
    if not density:
        plt.ylabel('Frequency')
    remove_border()
    plt.show()
    return
```
